B: Horner's Algorithm

```python
# initialize p = a_n; n = 10; a_i = 1/i
m = 10
x_val = [1, 5, 10]
for x in x_val: 
  p = 1/(m-0)
  for i in range(1, m+1):
    if i == m:
      p = p*x + 0 # a_0 = 0
    else: 
      p = p*x + 1/(m-i)
  print("When x =", x, ", the value of the polynomial is", p)
```
