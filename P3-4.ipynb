### Bisection Method

```python
def Bisect(func, M, width, tolerance, a, b):
  u = func(a)
  v = func(b)
  if u*v > 0:
     print("interval [" + str(a) + ", " + str(b) + "] doesn't contain 0.")
     return
  e = b - a
  for k in range(1, M+1):
    e = e / 2
    c = a + e
    w = func(c)
    if abs(e) < width or abs(w) < tolerance:
      print("find the root c:", c)
      print("f(c) = ", w)
      return k
    if u*w > 0:
      a = c
      u = w
    else:
      b = c
      v = w
  print("find the root c:", c)
  print("f(c) = ", w)
```

```python
def f1(x):
  return 1/x - 2**x

def f2(x):
  return x - math.tan(x)
```

```python
Bisect(f1, 10000, 10**-16, 10**-24, 10**-4, 1)
```

```python
Bisect(f2, 10000, 10**-16, 10**-24, 1, 2)
```

### Newton's Method

```python
from sympy import *
import math
import numpy as np
```

```python
# F is m*1, X is m*1
def Newton(M, tolerance, step, x_0, func, j_func):
  X = x_0
  F = func(X)
  for i in range(1, M+1):
    J = j_func(X) # m*m 
    if len(X) == 1:
      h = -1/J*F
    else:
      # solve the system J*h = -F(X)
      h = np.linalg.solve(J, -F)
    for k in range(0, len(X)):
      X[k] = X[k] + h[k]
    F = func(X)

    print("distance to the true root: ", np.linalg.norm(F))
    print("update h: ", np.linalg.norm(h))
    print()
    
    if np.linalg.norm(F) < tolerance or np.linalg.norm(h) < step:
      print("find the root X: ", X)
      print("F(X) =", F)
      return i
  print("find the root X: ", X)
  print("F(X) =", F)
```

### Secant's Method

```python
import math
```

```python
def swap(a, b):
  a = a + b
  b = a - b
  a = a - b
  return a, b
```

```python
# 
def secant(func, a, b, M, delta, epsilon):
  f_a = func(a)
  f_b = func(b)
  for k in range(2, M+1):
    if abs(f_a) > abs(f_b):
      f_a, f_b = swap(f_a, f_b)
      a, b = swap(a, b)
    s = (a - b) / (f_a - f_b)
    b = a
    f_b = f_a
    a = a - s*f_a
    f_a = func(a)
    
    print("abs(f(a)) = ", abs(f_a))
    print("abs(a-b) = ", abs(a-b))
    print()

    if abs(f_a) < epsilon or abs(a-b) < delta:
      print("found a:", a)
      print("func(a) =", f_a)
      return k-1
      break
  print("found a:", a)
  print("func(a) =", f_a)
  return M
```

```python
def f1(x):
  return math.sin(x/2)-1
```

```python
secant(f1, 0, 1, 1000, 10**-5, 10**-5)
```
